```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Frailty_Raw.csv')
df
```

```python
print("Shape:", df.shape, "\n")
df.dtypes
```

```python
df.columns=['Height', 'Weight', 'Age', 'Grip Strength','Frailty']
df.columns
df['Frailty'].replace(['Y', 'N'], [1, 0], inplace=True)
df
```

```python
df.to_csv('Clean_Frailty.csv')
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Height'], y=df['Weight'], data=df, hue='Frailty', palette='rocket_r')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatter Plot of Height vs. Weight')
plt.grid(True)
plt.legend(title='Frailty')
plt.show()
```

```python
sns.pairplot(df, hue="Frailty")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Frailty', data=df)
plt.title('Frailty Distribution')
plt.show()
```
