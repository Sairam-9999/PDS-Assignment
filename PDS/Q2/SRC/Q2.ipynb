```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv( 'StudentsPerformance.csv' )
```

```python
df
```

```python
df['lunch'].replace(['standard', 'free/reduced'], [0, 1], inplace=True)
df['test preparation course'].replace(['none', 'completed'], [0, 1], inplace=True)

df['Total score'] = df['math score'] + df['writing score'] + df['reading score']
df
```

```python
df.to_csv('Clean_SP.csv')
```

```python
# Pie Chart of Parental level of Education Types
a = df['parental level of education'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(a, labels=a.index, autopct='%1.1f%%', colors=['blue', 'yellow', 'red', 'grey', 'green', 'orange'])
plt.title('Pie Chart of Parental level of Education Types')
plt.show()
```

```python
# Box Plot of Math Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='Total score', data=df, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Total score')
plt.title('Box Plot of Total score by Gender')
plt.grid(axis='y')
plt.show()
```

```python
# Histogram of reading Scores
plt.figure(figsize=(8, 6))
sns.histplot(df['reading score'], bins=20, kde=True, color='tan')
plt.xlabel('reading Score')
plt.ylabel('Frequency')
plt.title('Histogram of reading Scores')
plt.grid(True)
plt.show()
```

```python
# Bar Chart of test preparation course Types
plt.figure(figsize=(10, 6))
sns.countplot(x='test preparation course', data=df, palette='rocket', order=df['test preparation course'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.xlabel('test preparation course')
plt.ylabel('Count')
plt.title('Bar Chart of test preparation course Types')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter Plot of Writing vs. Math Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='writing score', y='math score', data=df, hue='gender', palette='mako', alpha=0.9)
plt.xlabel('writing score ')
plt.ylabel('math Score')
plt.title('Scatter Plot of Writing vs. Math Scores')
plt.grid(True)
plt.legend(title='Gender')
plt.show()
```
